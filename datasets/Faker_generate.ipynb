```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from random import uniform 
from faker import Factory,Faker
from sklearn.preprocessing import normalize,MinMaxScaler
from pandas import to_datetime
from random import randint
import datetime as dt
%matplotlib inline

fake = Faker()
```

### 2. Build the real estate data table

```python
import random

def random_coord(start, end):
    return round(random.uniform(start, end), 6)

real_estate_data = []
for i in range(100000):
    state = fake.state()
    city = fake.city()
    street_address = fake.street_address()
    real_estate = {
        'real_estate_id': i + 1,
        'real_estate_owner': fake.company(),
        'street_address': street_address,
        'business_name': fake.company(),
        'state': state,
        'zip_code': fake.zipcode(),
        'num_investors': fake.random_int(min=1, max=100),
        'cost': fake.random_int(min=100000, max=10000000) / 100,
        'building_type': fake.random_element(elements=('Residential', 'Commercial', 'Industrial')),
        'building_material': fake.random_element(elements=('Brick', 'Concrete', 'Steel', 'Wood')),
        'insured_value': fake.random_int(min=100000, max=10000000) / 100,
        'city' : city,
        'latitude' : random_coord(24.396308, 49.384358),
        'longitude' : random_coord(-125.000000, -66.934570)
    }
    real_estate_data.append(real_estate)

# Create a pandas DataFrame from the real estate data
real_estate_df = pd.DataFrame(real_estate_data)

real_estate_df.reset_index(inplace = True, drop = True)

real_estate_df.head(5)
```
